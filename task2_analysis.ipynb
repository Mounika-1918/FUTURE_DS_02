```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('/content/customer_churn_dataset-testing-master.csv')
print("Shape:", df.shape)
print("\nColumn Names:")
print(df.columns.tolist())
print("\nFirst 5 rows:")
df.head()
```

```python
print("Missing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:", df.duplicated().sum())

print("\nData Types:")
print(df.dtypes)

# Basic statistics
print("\nBasic Statistics:")
df.describe()
```

```python
# 1. Overall Churn Rate
churn_rate = df['Churn'].value_counts()
churn_pct = df['Churn'].value_counts(normalize=True) * 100
print("Churn Count:")
print(churn_rate)
print("\nChurn Percentage:")
print(churn_pct)

# 2. Churn by Subscription Type
sub_churn = df.groupby('Subscription Type')['Churn'].mean() * 100
print("\nChurn Rate by Subscription Type:")
print(sub_churn)

# 3. Churn by Contract Length
contract_churn = df.groupby('Contract Length')['Churn'].mean() * 100
print("\nChurn Rate by Contract Length:")
print(contract_churn)

# 4. Churn by Gender
gender_churn = df.groupby('Gender')['Churn'].mean() * 100
print("\nChurn Rate by Gender:")
print(gender_churn)

# 5. Average Tenure of Churned vs Retained
tenure_churn = df.groupby('Churn')['Tenure'].mean()
print("\nAverage Tenure (0=Retained, 1=Churned):")
print(tenure_churn)
```

```python
# 1. Churn by Support Calls
support_churn = df.groupby('Support Calls')['Churn'].mean() * 100
print("Churn Rate by Support Calls:")
print(support_churn)

# 2. Churn by Payment Delay
df['Payment_Delay_Group'] = pd.cut(df['Payment Delay'],
                                    bins=[0,10,20,30],
                                    labels=['Low(0-10)','Medium(10-20)','High(20-30)'])
delay_churn = df.groupby('Payment_Delay_Group')['Churn'].mean() * 100
print("\nChurn Rate by Payment Delay:")
print(delay_churn)

# 3. Churn by Age Group
df['Age_Group'] = pd.cut(df['Age'],
                          bins=[18,30,45,65],
                          labels=['Young(18-30)','Middle(30-45)','Senior(45-65)'])
age_churn = df.groupby('Age_Group')['Churn'].mean() * 100
print("\nChurn Rate by Age Group:")
print(age_churn)

# 4. Revenue Lost due to Churn
churned_revenue = df[df['Churn']==1]['Total Spend'].sum()
retained_revenue = df[df['Churn']==0]['Total Spend'].sum()
print(f"\nRevenue Lost (Churned): ${churned_revenue:,.0f}")
print(f"Revenue Retained: ${retained_revenue:,.0f}")
```

```python
# Cohort Analysis - Tenure Groups
df['Tenure_Group'] = pd.cut(df['Tenure'],
                    bins=[0,12,24,36,48,60],
                    labels=['0-12 months','13-24 months',
                            '25-36 months','37-48 months',
                            '49-60 months'])

cohort_churn = df.groupby('Tenure_Group')['Churn'].mean() * 100
print("Churn Rate by Tenure Group:")
print(cohort_churn)

# Retention Rate
cohort_retention = 100 - cohort_churn
print("\nRetention Rate by Tenure Group:")
print(cohort_retention)

# Customer Lifetime Value
clv = df.groupby('Churn')['Total Spend'].agg(['mean','sum','count'])
print("\nCustomer Lifetime Value:")
print(clv)
```

# Customer Retention & Churn Analysis
## Dataset: 64,374 Customer Records

---

## 📊 Key Findings

### 1. Overall Churn Rate
- Total Customers: 64,374
- Churned: 30,493 (47.4%)
- Retained: 33,881 (52.6%)
- **Insight:** Almost half the customers are leaving — urgent action needed!

---

### 2. Churn by Contract Length
- Monthly Plan: 51.6% churn (Highest!)
- Annual Plan: 46.2% churn
- Quarterly Plan: 44.0% churn (Best retention!)
- **Insight:** Promote Quarterly & Annual plans to reduce churn!

---

### 3. Churn by Payment Delay
- Low Delay (0-10 days): 10.1% churn
- Medium Delay (10-20 days): 38.7% churn
- High Delay (20-30 days): 76.6% churn
- **Insight:** Payment delay is the #1 churn predictor!

---

### 4. Churn by Support Calls
- 0-4 calls: ~23-31% churn
- 5+ calls: ~60% churn
- **Insight:** Poor support experience drives customers away!

---

### 5. Churn by Age Group
- Young (18-30): 44.4% churn
- Middle (30-45): 44.7% churn
- Senior (45-65): 51.0% churn
- **Insight:** Senior customers need special attention!

---

### 6. Cohort Analysis — Tenure Groups
- 0-24 months: ~31% churn
- 25-60 months: ~56% churn
- **Insight:** 2-year mark is critical — loyalty programs needed!

---

### 7. Revenue Impact
- Revenue Lost (Churned): $15,836,117
- Revenue Retained: $18,991,722
- **Insight:** Reducing churn = direct revenue growth!

---

## 🎯 Top Business Recommendations

1. **Fix Payment Issues** — Auto-pay reminders & easy payment options
2. **Improve Support** — Faster response, better FAQ
3. **2-Year Loyalty Program** — Special offers at 24-month mark
4. **Senior-Friendly UX** — Dedicated support for 45+ age group
5. **Promote Quarterly Plans** — Better retention than monthly plans
